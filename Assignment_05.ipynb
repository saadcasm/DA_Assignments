```python
def count_vowels(string):
    vowels = "aeiou"
    vowel_count = 0
    
    for char in string.lower():
        if char in vowels:
            vowel_count += 1
    
    if vowel_count < 2:
        return f"{vowel_count} vowel"
    else:
        return f"{vowel_count} vowels"

print(count_vowels("Saad Sadiq"))
```

```python
def currency_converter(amount, exchange_rate = 280):

    # We are assuming that the user here wants to convert to PKR that's why we are mentioning PKR in the message
    return f"{amount} USD with an exchange rate of {exchange_rate} = {amount * exchange_rate} PKR"

print(currency_converter(500))
print(currency_converter(100, 300))
```

```python
def display_student_info(*, name, age, grade, school):
    print(f"The name of the student is {name}")
    print(f"The age of the student is {age}")
    print(f"The grade of the student is {grade}")
    print(f"The school of the student is {school}")

display_student_info(name = "Saad", age = 24, grade = "A", school = "DHA Suffa University")

#display_student_info(fullname = "saad", age = 24, score = "A", university = "DHA Suffa University") this will give an error as you cannot change the name of arguments here
#display_student_info("saad", 24, "A", "DHA Suffa University") this will give an error as it is necessary to provide keyword args
```

```python
def calculate_price(price, quantity):
    return int(price) * int(quantity)

print(f"The total price of the product is {calculate_price(20, 80)}")
```

```python
def display_prices(product_name, **product_prices):
    print(f"• Product: {product_name}")
    
    for store, price in product_prices.items():
        print(f"• {store}: ${price}")

display_prices("Laptop", Amazon = 800, eBay = 750, Walmart = 780)
```

```python
def find_max(*numbers):
    if len(numbers) != 0:
        return max(numbers)
    else:
        return "No numbers provided"

print(find_max(10, 20, 300, 40, 50, 100))
```

```python
def display_info(name, age, **other_details):
    print(f"• Name: {name}")
    print(f"• Age: {age}")
    print("• Other Details:")

    for key, value in other_details.items():
            print(f"\t- {key}: {value}")

display_info("Saad", 24, address = "123 Main St.", phone = "123-456-789", email = "saadsadiq2000@gmail.com")
    
```

```python
def calculate_rectangle(length, width):
    
    def calculate_area(length, width):
        return length * width

    def calculate_perimeter(length, width):
        return 2 * (length + width)

    return calculate_area(length, width), calculate_perimeter(length, width)

rectangle_area_perimeter = calculate_rectangle(20, 10)

print(f"Area of the Rectangle: {rectangle_area_perimeter[0]}")
print(f"Perimeter of the Rectangle: {rectangle_area_perimeter[1]}")
        
```

```python
products = [ {"name": "Laptop", "price": 1200, "in_stock": True, "discount": 0.1},
            {"name": "Smartphone", "price": 800, "in_stock": False, "discount": 0.05},
            {"name": "Headphones", "price": 150, "in_stock": True, "discount": 0.2},
            {"name": "Smartwatch", "price": 300, "in_stock": True, "discount": 0},
            {"name": "Keyboard", "price": 100, "in_stock": True, "discount": 0.15} ]

in_stock_products = [product["name"] for product in products if product["in_stock"] == True]
discounted_prices = [(product["name"], product["price"] * (1 - product["discount"])) for product in products if product["in_stock"] == True]

print(f"In-stock product names: {in_stock_products}")
print(f"Discounted prices: {discounted_prices}")
```
