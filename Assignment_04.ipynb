```python
my_tuple = ((1,2,3), [4,5,6], ('apple', 'banana', 'cherry'), [7,8,9])

for element in my_tuple:
    print(element)

first_element = (my_tuple[0])
last_element = (my_tuple[-1])
new_tuple = first_element + tuple(last_element)

print(f"\nNew Tuple: {new_tuple}")
```

```python
inventory = {"The Great Gatsby": {"price": 10.99, "quantity": 5}, 
             "Rich dad poor dad": {"price": 8.99, "quantity": 8}, 
             "Harry Potter": {"price": 12.50, "quantity": 2}}

inventory.update({"To Kill a Mockingbird": {"price": 10.00, "quantity": 0}})

print(inventory)

books_to_delete = []

for key, value in inventory.items():
        if value["quantity"] < 1:
            books_to_delete.append(key)

for book in books_to_delete:
    del inventory[book]

print("\n")
print(inventory)
```

```python
course_a = {"John", "Alice", "Bob", "David"}
course_b = {"Alice", "Eve", "Charlie", "David"}

students_in_course_a = set(course_a) - set(course_b)
print(f"Students who are in Course a but not in Course b: {students_in_course_a}")

students_in_one_course = set(course_a) ^ set(course_b)
print(f"Students who are only in one of the two courses: {students_in_one_course}")
```

```python
sum_of_even_numbers = 0

for i in range(1, 51):
    if i % 2 == 0:
        sum_of_even_numbers += i

print(f"Sum of all the even number from 1 to 50 is {sum_of_even_numbers}")
```

```python
number = int(input("Please input a number between 1 and 10: "))
print("\n")

for i in range(1, 11):
    print(f"{number} x {i} = {number * i}")
```

```python
number = int(input("Please enter a number: "))
print("\n")

exponent = 1

while exponent <= 10:
    print(f"{number}^{exponent} = {number ** exponent}")
    exponent += 1
```

```python
number = int(input("Enter a number to start the countdown: "))
print("\n")

while number != -1:
    print(number)
    number -= 1

print("Countdown finished!")
```
